- Series=one dimensional labeled array pd.series(data)

```python
# import the series
import pandas as pd
a=[2,3,4,5,6,7]
b=pd.Series(a)
b
```

```python
print(type(b))
```

```python
# get the element
print(b[2])
```

```python
# change the index num.
a=[2,3,4,5,6,7]
b=pd.Series(a,index=['a','b','c','d','e','e'])
b
```

```python
# datatype changeing
a=[2,3,4,5,6,7]
c=pd.Series(a,dtype="float")
c
```

```python
dict={
    "name":["python","java","c","c++"],
    "por":[12,13,14,15],
    "rank":[1,4,5,3]
}
var=pd.Series(dict)
var
```

```python
a=pd.Series(12,index=[1,2,3,4,5,6,7,8])
b=pd.Series(12,index=[2,3,4,5])
print(a+b)
```
